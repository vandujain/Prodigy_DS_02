
EDA

```python
# Age distribution
plt.figure(figsize=(6,3))
sns.histplot(dd["Age"],kde=True)
plt.title("Age Distributed")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Survival rate by gender
plt.figure(figsize=(6,3))
sns.countplot(data=dd,x="Sex",hue="Sex")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived",loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(data=dd,x="Age",y="Fare", hue="Age")
plt.title("Scatter plot of Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```

```python
# Survival based on passenger class
plt.figure(figsize=(6,3))
sns.barplot(x='Pclass', y='Survived', data=dd)
plt.title("Bar plot of Pclass and Survived")
plt.xlabel("Pclass")
plt.ylabel("Survived")
plt.show()
```

**Correlation between Variables**
